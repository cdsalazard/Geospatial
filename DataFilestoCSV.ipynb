```python
import pandas as pd
import re
import os
```

```python
#Leamos los nombres de todos los archivos y los guardamos en un array
carpeta = 'PrecipitacionNacionalDiaria'
with os.scandir(carpeta) as ficheros:
    ficheros = [fichero.name for fichero in ficheros if fichero.is_file() and fichero.name.endswith('.data')]
ficheros
```

```python
#Leemos y agrupamos los datos por a침o y por mes y calculamos el promedio de lluvia en ese mes

# Creamos el dataset vacio en el cual vamos a guardar todos los valores
df = pd.DataFrame(columns =['Estacion', 'Year', 'Mes', 'Dia', 'Ubicacion_x', 'Valor'] )

for fichero in ficheros:
    nombre = re.findall('\d+' ,fichero)  #Este es el nombre de la estaci칩n
    nombre
    #Leemos los datos, creamos las columnas para la estaci칩n, el a침o, el mes y calculamos el promedio
    data =  pd.read_csv("PrecipitacionNacionalDiaria/" + fichero, sep = "|")
    data['Year'] = data.Fecha.str.split(pat='-',expand=True)[0]
    data['Mes'] = data.Fecha.str.split(pat='-',expand=True)[1]
    data.drop(['Fecha'], axis = 1, inplace = True)
    agrupado = data.groupby(by = ['Year', 'Mes'])['Valor'].mean()
    agrupado = agrupado.to_frame().reset_index()
    agrupado.insert(0, 'Estacion', nombre[0])
    agrupado
    df = df.append(agrupado, ignore_index=True)
df
```

```python
df.to_csv('Mensual_todas_estaciones.csv')
```
